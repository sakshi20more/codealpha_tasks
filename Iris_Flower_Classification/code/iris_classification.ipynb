```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the training and testing datasets from the drive
train_df = pd.read_csv('/content/drive/MyDrive/Iris_classification_task/Data/iris_train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/Iris_classification_task/Data/iris_test.csv')
```

```python
train_df.head()
test_df.head()
```

```python
#Separating features and target
X_train = train_df.drop('Species', axis=1)
y_train = train_df['Species']
X_test = test_df.drop('Species', axis=1)
y_test = test_df['Species']
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Training the model using Logistic Regression and scikit learn
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
#predicting model
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test,y_predict))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
